æœ€å¾Œï¼Œè¼‰å…¥æ¨¡å‹ä¸¦åŸ·è¡Œä¸€æ¬¡ç°¡å–®çš„ç”Ÿæˆã€‚ (æ³¨æ„ï¼šç”±æ–¼ Wan2.1 æ˜¯æ¥µæ–°çš„å°ˆæ¡ˆï¼Œä»¥ä¸‹ä»£ç¢¼å‡è¨­å…¶æ¨™æº–èª¿ç”¨æ–¹å¼ï¼Œè‹¥å®˜æ–¹ API æœ‰è®Šå‹•ï¼Œéœ€åƒè€ƒå…¶ predict.py æˆ– inference.py)

````python
import sys
# ç¢ºä¿ç•¶å‰ç›®éŒ„åœ¨ Python è·¯å¾‘ä¸­
sys.path.append(os.getcwd())

# å˜—è©¦å°‹æ‰¾ä¸¦åŸ·è¡Œæ¨ç†è…³æœ¬
# æˆ‘å€‘ä½¿ç”¨ Python çš„ subprocess ä¾†åŸ·è¡Œå‘½ä»¤ï¼Œé€™æ¨£åœ¨ Windows ä¸Šæ¯”è¼ƒç©©å®š
import subprocess

# å®šç¾©æç¤ºè©
prompt = "A cinematic drone shot of a futuristic cyberpunk city at night, neon lights, rain, high quality, 4k"
output_file = "output_test.mp4"

# æ§‹å»ºå‘½ä»¤ (é€™æ¨¡æ“¬äº†æ‚¨åœ¨ CMD ä¸­è¼¸å…¥æŒ‡ä»¤)
# æ³¨æ„ï¼šé€™è£¡å‡è¨­å°ˆæ¡ˆä¸­æœ‰ä¸€å€‹ generate.py æˆ–é¡ä¼¼çš„ä¸»ç¨‹å¼
# å¦‚æœå°ˆæ¡ˆä½¿ç”¨çš„æ˜¯ wan_t2v.pyï¼Œè«‹å°‡ä¸‹æ–¹ 'generate.py' æ›¿æ›ç‚ºå¯¦éš›æª”å
script_name = "generate.py"

if not os.path.exists(script_name):
    # å˜—è©¦å°‹æ‰¾å¸¸è¦‹çš„æ¨ç†è…³æœ¬åç¨±
    possible_scripts = ["inference.py", "predict.py", "wan_t2v.py"]
    for s in possible_scripts:
        if os.path.exists(s):
            script_name = s
            break

print(f"å°‡ä½¿ç”¨ {script_name} é€²è¡Œç”Ÿæˆ...")

# åŸ·è¡Œç”ŸæˆæŒ‡ä»¤
# æ³¨æ„ï¼š--checkpoint_dir çš„åƒæ•¸åç¨±å¯èƒ½éœ€è¦æ ¹æ“šå¯¦éš›ä»£ç¢¼èª¿æ•´
cmd = [
    "python", script_name,
    "--task", "t2v-1.3b",
    "--checkpoint_dir", model_path,
    "--prompt", prompt,
    "--output_dir", "./results"
]

print(f"åŸ·è¡ŒæŒ‡ä»¤: {' '.join(cmd)}")

# é€é subprocess åŸ·è¡Œï¼Œé€™æ¨£å¯ä»¥çœ‹åˆ°å³æ™‚è¼¸å‡º
process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, encoding='utf-8')

# é¡¯ç¤ºåŸ·è¡Œéç¨‹
for line in process.stdout:
    print(line, end='')

process.wait()

if process.returncode == 0:
    print("\nâœ… å½±ç‰‡ç”Ÿæˆå®Œæˆï¼è«‹æª¢æŸ¥ results è³‡æ–™å¤¾ã€‚")
else:
    print("\nâŒ ç”Ÿæˆéç¨‹ä¸­ç™¼ç”ŸéŒ¯èª¤ã€‚")
````
